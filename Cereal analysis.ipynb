```python
# pair plot for data for overall distribution of data:

sns.pairplot(df)
plt.show()
```

```python
# now we will rename some columns name to understand properly what the dataset what to make us know :

df.rename(columns={'mfr':'manufacturer','protein':'protein(g)','fat':'fat(g)','sodium':'sodium(mg)','fiber':'fiber(g)','carbo':'carbohydrates(g)','sugars':'sugars(g)','potass':'potassium(mg)','vitamins':'vitamins%','weight':'weight(oz/serv)','cups':'cups/serv'},inplace=True)
```

```python
# now we will replace all manufacturers name from single alphabet to proper name to understand manufacturer name easily :

df['manufacturer']=df['manufacturer'].replace({'A':'American','G':'General mills','K':'Kelloggs','N':'Nabisco','P':'Post','Q':'Quaker Oats','R':'Ralston Purina'})
```

```python
# now we will replace type name as :

df['type']=df['type'].replace({'C':'Cold','H':'Hot'})
```

```python
# lets see our new dataset :

df.head()
```

# Visualization, Univariate and Bivariate :

```python
# lets visualizing correlation of various elementswith each other :

plt.figure(figsize=(15,10))
hm=sns.heatmap(data=df.corr(),annot=True,square=True)
plt.show()

```

```python
num_vars=['calories','protein(g)','fat(g)','sodium(mg)','fiber(g)','carbohydrates(g)','sugars(g)','potassium(mg)','vitamins%','shelf','weight(oz/serv)','cups/serv','rating']
```

```python
cat_vars=['name','manufacturer','type']
```

```python
#  Univariate Analysis :
# Build histplot for numerical data :

for i in num_vars:
    sns.histplot(x=df[i])
    plt.show()
```

```python
# build countplot for categorical variables :

for i in cat_vars:
    sns.countplot (x=df[i])
    plt.show()
```

```python
# from above we can see hot type is heavily underdemand or under sampled..
```

```python
##  Bivariate Analysis :

# for numerical variable :

sns.pairplot(df[num_vars],diag_kind='kde')
plt.show()
```

```python
#  for categorical variables :

for i in cat_vars:
    sns.boxplot(x=df[i],y=df['rating'])
    plt.show()
```
